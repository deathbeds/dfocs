# your files as data

`dfocs` represents you files as data. the `Frames`

```python
    from IPython.display import display
    from dfocs.docs import Docs, Config
```

```python
    self = Docs(Config(src="../.."))
```

the initialize step seeds the files and cells to operate on. all files get represented with the notebook schema because it is generic enough.

```python
    self.initialize()
    display(self.files, self.cells)
```

the expand step adds more columns to the cells frame. metadata is extracted from the rows and modifications are made to the dataframe contents.

```python
    self.expand()
    display(self.files, self.cells)
```

compaction passes cell level information to the files. for example, title and description are passed to the files for use in the head of the document.

```python
self.compact()
display(self.files, self.cells)
```
